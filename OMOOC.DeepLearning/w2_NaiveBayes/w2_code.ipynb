## 线性回归

```python
%matplotlib inline
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

### 生成一些有随机干扰的数据  

$y = x + noise$

```python
size = 100
x = np.arange(size)
y = x + np.random.normal(scale=5, size=size)
```

```python
plt.plot(x,y,'.');
```

### 用线性函数拟合这些数据

```python
def linear_function(x, w, b):
    return w * x + b

plt.plot(x, y, '.')
plt.plot(x, linear_function(x, 1, 0));
```

### 判定拟合的好坏

```python
np.arange(0, 3, 0.1)
```

```python
b = 0
all_costs = []
for w in np.arange(0, 3, 0.1):
    y_bar = linear_function(x, w, b)
    cost = ((y_bar - y)**2).sum()
    all_costs.append((w, cost))

all_costs = np.asarray(all_costs)
```

```python
type(all_costs)
```

```python
plt.plot(all_costs[:,0], all_costs[:,1]);
```

### 变量变成两个的情况

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
np.random.normal(scale=2, size=10) #  scale 标准差
```

```python
np.random.randint(0,3, (5,2)) # 0-3 5*2 随机矩阵
```

```python
size = 100
x = np.random.randint(0, 100, (size, 2))
x1 = x[:, 0]
x2 = x[:, 1]
y = x[:, 0] + x[:, 1] + np.random.normal(scale=2, size=size)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1, x2, y);
```

```python
def linear_function(x1, w1, x2, w2, b):
    return x1 * w1 + x2 * w2 +b

b = 0
all_costs = []
for w1 in np.arange(0, 3, 0.1):
    for w2 in np.arange(0, 3, 0.1):
        y_bar = linear_function(x[:,0], w1, x[:,1], w2, b)
        cost = ((y_bar - y)**2).sum()
        all_costs.append((w1, w2, cost))

all_costs = np.asarray(all_costs)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(all_costs[:, 0], all_costs[:, 1], all_costs[:, 2]);
```
